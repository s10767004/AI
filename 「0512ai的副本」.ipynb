```python
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from time import time
import numpy as np
import matplotlib.pyplot as plt

from sklearn import metrics

X= np.array([[1,1],[1.1,1.1],[1.2,1.2],[2,2],[2.1,2.1],[2.2,2.2]])
y= [1,1,1,
   0,0,0]
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
print("集群中心的座標",kmeans.cluster_centers_)
print("預測:",kmeans.predict(X))
print("實際:",y)
print("預測[1,1],[2.3,2.1]:",kmeans.predict([[1, 1],[2.3,2.1]]))
plt.axis([0, 3, 0, 3])
plt.plot(X[:3,0],X[:3,1], 'yx')
plt.plot(X[3:,0],X[3:,1], 'g.')
plt.plot(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], 'ro')
plt.xticks(())
plt.yticks(())
plt.show()
```

1.kmeans是最簡單的聚類算法之一，但是運用十分廣泛。選取適當的k，將數據分類後，然後分類研究不同聚類下數據的特點。舉凡客戶分類、文檔分類等等，尤其是在某特別類型的抓取有著不錯的表現。
**優點**:計算速度快，且易理解。
**缺點**:平均值的算法易受極端值得影響，且對於類別型變數來講需考慮計算平均值並無意義，故只適用於數值型的資料。另外，當樣本數量差異過大時，也較無法達到良好的分類效果。

2.至於決策樹是一種解決分類問題的算法，是一種樹形結構。決策樹利用層層推理來實現數據的分類，決策樹的構成可分為：根節點(包含全部的數據樣本)、內部節點(對應的特徵屬性測試)、葉節點(決策的結果)。每個節點的部分，都是一個特徵。根據特徵一步又一步的細分，將得到分類結果。
**優點**:易於理解和解釋，也可以同時處理標稱稱和和數值型數據，它比較適合處理有缺失屬性的樣本，且在相對短的時間能夠對大型數據源取得可行且效果良好的結果。
**缺點**:容易發生過擬合，也易忽略數據集中屬性的相互關聯。對於那些那些類別樣本數量不計其數的數據，在決策樹中，進行屬性劃分時，不同的判定規則會帶來不同的屬性選擇變量
